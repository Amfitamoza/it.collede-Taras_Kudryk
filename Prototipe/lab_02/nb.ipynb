## Загальні поняття
>  клас - це загальний темплейт який описує щось конкретне;
> обєкт - це окремий екземпляр класу з своїми характеристиками;
> тип даних - це характеристика набору даних;

- Вбудовані типи даних (текстові/стрічкові)

```python
print("Це текстові дані які будуть виведені")
a = "Ця змінна містить текстові дані"
print(a)
b = str("Це будуть текстові дані через клас str")
print(b)

print("Рахуємо скільи раз зустрічається буква 'е' в обєкті b: ", b.count("у"))
print("Рахуємо скільи раз зустрічається буква 'і' в текстовій зміні а: ", a.count("і"))
print("Рахуємо скільки раз зустрічається буква 'і' в простій стрічці", "Це буде стрічка без присвоєння змінній чи створення обєкту" .count("і"))
```
