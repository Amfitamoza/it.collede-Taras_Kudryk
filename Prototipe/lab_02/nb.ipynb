- Запиталиcь в ChatGPT як вивести піраміду з смайликів

> Для створення піраміди з символів ": )" можна використовувати цикли. Ось приклад коду на Python, який буде виводити піраміду, починаючи з вашого початкового рядка:

```python
def print_smiley_pyramid(rows):
    for i in range(1, rows + 1):
        print(' ' * (rows - i) + ':)' * (2 * i - 1))

# Кількість рядків у піраміді
num_rows = 8

print_smiley_pyramid(num_rows)
```

- Робота з циклами

>`while` - умовний цикл - ітерації будуть виконуватись поки умова є істинною(може бути безмежним циклом)

>`for` - з заданою кількістю інтерацій (повторів)

```python
from random import randint
n = ["Тарас", "Адольф", "Степко"]
l = ["стрільбу", "Бахмут", "мілітаризм"]

j = 0
while j < 3:
    print(">>> {name} любить {love}\n".format(name=n[j],love=l[randint(0,2)]))
    j+=1

print("="*10,"Це вже не входить в цикл, цикл закінчився. Цикл while не дуже підходить для визначення кількості інтерацій")

# навіть для for можна йти по різному
for j in range(len(n)):
    print(">>> {name} любить {love}\n".format(name=n[j],love=l[randint(0,2)]))
    
print("="*10,"Це вже виконуєтьсся поза циклом")

print(f"Демонструємо що буде виконувати функція range:, {list(range(3))}")

#варіант 2 для for (це вже правильне використання for)
for j in n:
    print(">>> {name} любить {love}\n".format(name=j,love=l[randint(0,2)]))
```

> Пробуємовикористати цикли при симуляції кидання кубика

```python
# Числа можуть бути тільки від 1 до 5, і ми їх вводимо з клавіатури
required_number = int(input("Введіть бажане число на кубику (від 1 до 6):"))

dice_number = int()
print(f"Значення яке повинно випасти на кубику відповідає умові від 1 до 6?")
if 1 <= required_number <=6:
    print("Введено коректні дані, починаємо кидати кубик:")
    while required_number != dice_number: #Задаємо умову неістинності
        dice_number = randint(1,6)
        print(f"Кидаємо кубик, випало значення: {dice_number}. Отримане значення по відношенню до бажаного буде: {required_number == dice_number}.")
else:
    print(f"Введемо не коректне значення: {required_number}")
```

```python
attempts = 4 # кількість спроб
required_number = 6 # число яке має випасти

dice_number = int()

# Якщо ми не будемо використовувати нашу ітераційну змінну у цмклі, то можемо її назвати _ (просто нижнє підкреслення)

for _ in range(attempts):
    dice_number = randint(1,6)
    print(f"Кидаємо кубик, випадає число {dice_number}. Отримаємо значення по відношенню до бажаного буде {required_number == dice_number}.")
    if required_number == dice_number:
        print(f"Умова виконана на {_ + 1} ітерації, виходимо з циклу.")
        break
else:
    print("Задане значення не було отримано. Ви програли.")
```

- пробуємо one-line (використання синтаксису if або for у одному рядку)
> може бути використано для простих умов які можна помістити в один рядок

```python
name = "Тарас"

# Канонічний вигляд

if name in n:
    print(f"Імя {name} значення n = {n}")
else:
    print(f"Немає імені {name} у списку, додаємо його: {n.append(name)}")

# Запис в один рядок

# print(f"Імя {name} знаходимо в списку n = {n}" if name in n else f"Немає імені {name} у списку.")
print(f"Імя {name} знаходимо в списку n = {n}" if name in n else f"Немає імені {name} у списку, додаємо його {n.append(name)}")

print(f"Кінцевий вигляд списку буде {n}")
```

```python
pn = list()
# канонічний запис циклу
for name in n:
    pn.append(f"Імя {name} має {len(name)} букв")

print(pn)

# переписуємо простий цикл в One liner
pn2 = list([f"Імя {name} має {len(name)} букв" for name in n])

print(pn2)
```

```python
from time import time
n = ["Тарас", "Адольф", "Кумис", "Баба Яга"]
print(f"Початковий вигляд списку: {n}")

# Це варіант використання for i if
start = time()
pn = list()
for name in n:
    if len(name) == 5:
        pn.append(f"Це {name}!")
    else:
        pn.append(f"Оце так імя {name}!")
end = time()
print(pn, f"\nСписок заповнювався {end - start} секунд")

# варіант використання for та if та модифікація за допомогою continue
pn1 = list()
for name in n:
    if len(name) == 5:
        pn1.append(f"Це {name}!")
        continue
    pn1.append(f"Оце так імя {name}!")  
print(pn1)

pn2 = list([f"Це {name}!" if len(name) == 5 else f"Оце так імя {name}!" for name in n])
print(pn2)
```
