- контактонація (додавання/зчеплення) та множення стрічок

```python
c = a + "| Це буде рядок зчеплення |" + b
print(c)

d = a.__add__("| Це будерядок зчеплення |".__add__(b))

print("В обох випадках будемо мати\n", c, "\n", d)

#Ці два записи рівноціні
e = a * 4
f = a.__mul__(4)
print("Що буде при множенні стрічки на число?:\n", e, "\n", f)

#
print("-_-_-_-_-" * 20)
print(" " * 5)
```

- Запиталиcm в ChatGPT як вивести піраміде з смайликів

> Для створення піраміди з символів "!" можна використовувати цикли. Ось приклад коду на Python, який буде виводити піраміду, починаючи з вашого початкового рядка:

```python
def print_smiley_pyramid(rows):
    for i in range(1, rows + 1):
        print(' ' * (rows - i) + ':)' * (2 * i - 1))

# Кількість рядків у піраміді
num_rows = 8

print_smiley_pyramid(num_rows)
```
